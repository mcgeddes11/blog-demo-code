```python
# You need wordcloud and matplotlib libraries to run this notebook
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from matplotlib import pyplot as plt
% matplotlib inline
```

```python
textfile = "/media/sf_VBoxShared/blog_posts_20190327.txt"
with open(textfile,"r") as f:
    post_content = f.read()
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS, background_color="white", width=640, height=480).generate(post_content)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("/media/sf_VBoxShared/blog_posts_20190327.png")
```
